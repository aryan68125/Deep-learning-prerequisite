```python
from sklearn.datasets import load_boston
import numpy as np
```

```python
boston = load_boston()
X = boston.data #data 
Y = boston.target #targets
```

```python
Y
```

```python
X.shape #(506 samples, 13 features)
```

```python
#Normalize or standardize our data
u = np.mean(X,axis=0)
std = np.std(X,axis=0)
X=(X-u)/std
```

```python
# here we will concatenate our column of ones with X to complete our algo
ones = np.ones((X.shape[0],1))
X = np.hstack((ones,X))
```

```python
X
```

```python
# here we will get our theta
#(X.TX)-1 X.TY
a = np.linalg.inv(np.dot(X.T,X))
b = np.dot(X.T,Y)
theta = np.dot(a,b)
theta
```

```python
theta.shape # theta will always be n+1 
```

```python
def hypothesis(X,theta):
    return np.dot(X,theta)
```

```python
ypred = hypothesis(X,theta)
```

```python
ypred
```

```python
def r2score(y,ypred):
    ymean = np.mean(y)
    num = np.sum((y-ypred)**2)
    demon = np.sum((y-ymean)**2)
    return 1-num/demon
```

```python
r2score(Y,ypred)
```
